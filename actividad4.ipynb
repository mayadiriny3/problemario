### *1. Completa la siguiente tabla de Representaciones de Fourier para cuatro clases de señales, según corresponda su relación entre filas y columnas.*

| Propiedades de tiempo  | Periodica| No periodica    |
|---------|------|--------------|
| Continua    |Una señal continua se caracteriza por tener valores definidos para todos los puntos en un intervalo de tiempo continuo. Esto significa que la señal puede tomar cualquier valor en un rango infinito de tiempo y no está limitada a valores discretos.   |En una señal continua no periódica, los valores de la señal pueden variar de manera arbitraria en cualquier instante de tiempo. Estas señales pueden presentar cambios suaves o bruscos, y no siguen un patrón predecible o repetitivo. |
| Discreta   |En una señal discreta periódica, la forma y los valores de la señal se repiten exactamente después de un cierto número de muestras. La periodicidad se expresa en términos de un período discreto (N), que indica cuántas muestras deben pasar para que la señal se repita.   |En una señal discreta no periódica, los valores de la señal pueden variar de manera arbitraria en cada punto discreto en el tiempo. Estas señales no siguen una secuencia repetitiva y pueden presentar cambios irregulares en sus valores a lo largo del tiempo discreto.|


```python
###CONTINUA PERIODICA
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de tiempo
t = np.linspace(0, 2*np.pi, 100)  # Genera 100 puntos en el rango de 0 a 2π

# Calcular los valores de x
x = np.cos(2*np.pi*t/np.pi)

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar la función
ax.plot(t, x)

# Configurar los ejes y etiquetas
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_title('Gráfica de x = cos(2πt/π)')

# Mostrar la gráfica
plt.show()
```

```python
### CONTINUA NO PERIODICA
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de tiempo t
t = np.linspace(-1, 2, 1000)  # Genera 1000 puntos en el rango de -1 a 2

# Definir la función x(t)
T = 1  # Período
x = np.zeros_like(t)
x[(t >= 0) & (t <= T)] = 1

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar la función
ax.plot(t, x)

# Configurar los ejes y etiquetas
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_title('Gráfica de x(t/T)')

# Mostrar la gráfica
plt.show()
```

```python
### DISCRETA PERIODICA
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de tiempo
t = np.linspace(0, 2*np.pi, 100)  # Genera 100 puntos en el rango de 0 a 2π

# Calcular los valores de x
x = 17 * np.cos(3*np.pi*t)

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar la función
ax.plot(t, x)

# Configurar los ejes y etiquetas
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_title('Gráfica de x = 17 * cos(3πt)')

# Mostrar la gráfica
plt.show()
```

```python
###DISCRETA NO PERIODICA
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de frecuencia Ω
Ω = np.linspace(-np.pi, np.pi, 100)  # Genera 100 puntos en el rango de -π a π

# Calcular los valores de X
X = np.exp(1j * Ω)

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar la función
ax.plot(Ω, X.real, label='Parte Real')
ax.plot(Ω, X.imag, label='Parte Imaginaria')

# Configurar los ejes y etiquetas
ax.set_xlabel('Ω')
ax.set_ylabel('X')
ax.set_title('Gráfica de X(e^(jΩt))')
ax.legend()

# Mostrar la gráfica
plt.show()
```
