6. Representa gráficamente los números complejos:    
a) (3+4i); b) -4 ; c) -2i; d) (-2+3i); e) (1+3i);f) (6-i) ; g) 3 ; h) 3i; j) (-1+i)


```python
import numpy as np
import matplotlib.pyplot as plt

# Configurar la convención de números complejos
np.set_printoptions(formatter={'complex_kind': lambda x: f"{x.real} + {x.imag}i"})

# Números complejos
numeros_complejos = [(3 + 4j), (-4), (-2j), (-2+3j), (1+3j), (6-1j), (3), (3j), (-1+1j)]

# Obtener las partes real e imaginaria
parte_real = [comp.real for comp in numeros_complejos]
parte_imaginaria = [comp.imag for comp in numeros_complejos]

# Crear el gráfico de dispersión con colores diferentes
plt.scatter(parte_real, parte_imaginaria, c=range(len(numeros_complejos)), cmap='rainbow')

# Configurar los ejes
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginaria')
plt.grid(True)

# Etiquetar los puntos
for i in range(len(numeros_complejos)):
    plt.annotate(f'({numeros_complejos[i].real} + {numeros_complejos[i].imag}i)', (parte_real[i], parte_imaginaria[i]), xytext=(5, 5), textcoords='offset points')

     
# Mostrar el gráfico
plt.show()
```

7. Efectuar los siguientes productos y expresa el resultado en forma polar y binómica:  
$a)(cos30º+isen30º)[2(cos15º+isen15º)]   
b) [2(cos23º+isen23º)] [3(cos37º+isen37º)]$

```python
import cmath
import math

# a) (cos30º+isen30º)[2(cos15º+isen15º)]
num_a = cmath.rect(1, math.radians(30))
num_b = cmath.rect(2, math.radians(15))
resultado_a = num_a * num_b
polar_a = cmath.polar(resultado_a)
binomica_a = resultado_a

print("Resultado a) en forma polar:", polar_a)
print("Resultado a) en forma binómica:", binomica_a)

# b) [2(cos23º+isen23º)] [3(cos37º+isen37º)]
num_c = cmath.rect(2, math.radians(23))
num_d = cmath.rect(3, math.radians(37))
resultado_b = num_c * num_d
polar_b = cmath.polar(resultado_b)
binomica_b = resultado_b

print("Resultado b) en forma polar:", polar_b)
print("Resultado b) en forma binómica:", binomica_b)
```
