```python
# Importar librerias basicas
import numpy as np
import sympy as sympy
import matplotlib.pyplot as plt
from sympy.abc import x, t, n
from scipy.signal import square
from sympy import *

# imprimir con notación matemática.
sympy.init_printing(use_latex='mathjax')

print("Bibliotecas cargadas")
```

## Sección 1


### Serie de Fourier


```python
#CALCULANDO SERIE 

s = fourier_series(x, (x, -pi, pi))
s.shift(1).truncate()
```

```python
#GRAFICANDO SERIE 

f = x
s = fourier_series(f, (x, -pi, pi))
s1 = s.truncate(n = 100)
p = plot(f, s1, (x, -pi, pi), show=False, legend=True)
p[0].line_color = '#000000'
p[0].label = 'x'
p[1].line_color = '#0097A7'
p[1].label = 'n=100'
p.show()
```

## Sección 2

```python
#CALCULANDO SERIE

from math import e

s = fourier_series(e**(-4*x), (x, -2, 2))
s.truncate(100)
```

```python
#GRAFICANDO SERIE 

from math import e
f = e**(-4*x)
s = fourier_series(f, (x, -2, 2))
s1 = s.truncate(n = 100)
p = plot(f, s1, (x, -2, 2), show=False, legend=True)
p[0].line_color = '#000000'
p[0].label = 'x'
p[1].line_color = '#0097A7'
p[1].label = 'n=100'
p.show()
```

```python
#CALCULO Y GRAFICACION DE SERIE 

ao = integrate(2 / pi, (t, 0, pi/2))
#integramos la funcion (2/pi) cuya variable es 't'
#y limites de integracion entre 0 y pi/2

#Se imprime "ao" 
print ("a0 = ")
ao
```

```python
#Integramos la funcion (2/pi)*cos(2nt)
an = integrate((2 / pi) * cos(2 * n * t), (t, 0, pi/2))
#Su variable es 't' y sus limites de integracion son 0 y pi/2

#Se imprime "an" 
print ("an = ")
an
```

```python
#integramos la funcion (2/pi*cos(2nt)
bn = together(integrate((2 / pi) * sin(2 * n * t), (t, 0, pi/2)))
#Su variable es 't' y sus limites de integracion
#son 0 y pi/2. Ademas usamos la funcion "together"
#para simplificar la expresion
print ("bn = ")
bn
```

```python
#Se imprime "serie" 

print ("f(x) = ")
armonicos = 100
serie = (ao/2)
for i in range(1, armonicos + 1):
    serie = serie + (an*cos(2*n*t)).subs(n, i)
for j in range(1, armonicos + 1):
    serie = serie + (bn*sin(2*n*t)).subs(n, j)

serie
```

```python
#Usando el modulo para graficas de sympy
plotting.plot(serie, ylim=(-0.5, 1.5), xlim=(-0.5,5), line_color = '#0097A7') 
plt.show()
```
